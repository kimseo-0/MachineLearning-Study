```python
from sklearn.preprocessing import LabelEncoder

def labelEncoder(df, features):
    for feature in features:
        le = LabelEncoder()
        le_result = le.fit(df[feature])
        df[feature] = le.transform(df[feature])
```

```python
def oneHotEncoder(df, features):
    df_dummies = df.copy()
    df_dummies = pd.get_dummies(df_dummies, columns=features, drop_first=True)
    return df_dummies
```

```python
def train_model(model, X_train, y_train):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def evaluate_regression(y_true, y_pred):
    print("MSE:", mean_squared_error(y_true, y_pred))
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("R²:", r2_score(y_true, y_pred))
```

```python
def random_forest_space(trial):
    params = {
        # n_estimators: 생성할 트리의 개수
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        
        # max_depth: 트리의 최대 깊이
        # 깊이의 제한이 없으면 과적합될 수 있습니다.
        # None은 제한이 없다는 의미로, 2에서 32 사이의 값으로 설정하는 경우가 많습니다.
        'max_depth': trial.suggest_int('max_depth', 2, 32, log=True),
        
        # min_samples_split: 노드를 분할하기 위한 최소 샘플 수
        # 작은 값은 과적합을 유발할 수 있습니다.
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        
        # min_samples_leaf: 리프 노드에 있어야 할 최소 샘플 수
        # 작은 값은 과적합을 유발할 수 있습니다.
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        
        # max_features: 각 분할에서 고려할 특성(feature)의 최대 개수
        # 'auto'는 `n_features`, 'sqrt'는 `sqrt(n_features)`를 의미합니다.
        # 이 옵션을 `trial.suggest_categorical`로 정의하면 좋습니다.
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None]),
    }
    return params
```

```python
def xgboost_space(trial):
    return {
        "n_estimators": trial.suggest_int("n_estimators", 100, 300),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "subsample": trial.suggest_float("subsample", 0.6, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.6, 1.0),
        "gamma": trial.suggest_float("gamma", 0, 5),
        "reg_lambda": trial.suggest_float("reg_lambda", 1e-3, 10),
        "reg_alpha": trial.suggest_float("reg_alpha", 1e-3, 10),
    }
```

```python
def lgbm_param_space(trial):
    """
    LightGBM 모델을 위한 하이퍼파라미터 탐색 공간을 정의합니다.
    """
    params = {
        # n_estimators: 부스팅 라운드(트리)의 수
        # 너무 크면 과적합될 수 있으므로, early stopping과 함께 사용하는 것이 좋습니다.
        'n_estimators': trial.suggest_int('n_estimators', 100, 2000, step=100),
        
        # learning_rate: 학습률
        # 작은 값은 더 많은 트리를 필요로 하지만 더 나은 일반화 성능을 제공할 수 있습니다.
        'learning_rate': trial.suggest_loguniform('learning_rate', 1e-3, 0.1),
        
        # num_leaves: 하나의 트리에 있는 최대 리프 노드 수
        # 트리의 복잡도를 제어하는 핵심 파라미터
        'num_leaves': trial.suggest_int('num_leaves', 8, 256),
        
        # max_depth: 트리의 최대 깊이
        # -1은 제한이 없음을 의미합니다. num_leaves가 우선순위를 가집니다.
        'max_depth': trial.suggest_int('max_depth', 3, 15),
        
        # min_child_samples: 리프 노드가 되기 위한 최소 샘플 수
        'min_child_samples': trial.suggest_int('min_child_samples', 20, 500, log=True),
        
        # feature_fraction: 트리를 만들 때마다 무작위로 선택할 특성(feature)의 비율
        # 과적합을 방지하는 데 도움
        'feature_fraction': trial.suggest_uniform('feature_fraction', 0.5, 1.0),
        
        # bagging_fraction: 트리를 만들 때마다 무작위로 선택할 데이터 샘플의 비율
        # 과적합을 방지하는 데 도움
        'bagging_fraction': trial.suggest_uniform('bagging_fraction', 0.5, 1.0),
        
        # bagging_freq: bagging을 수행할 빈도
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 7),
        
        # lambda_l1, lambda_l2: L1, L2 규제 파라미터
        # 너무 큰 값은 모델을 너무 단순하게 만들 수 있습니다.
        'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-8, 10.0),
        'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-8, 10.0)
    }
    return params
```

```python
def catboost_param_space(trial):
    """
    CatBoost 모델을 위한 하이퍼파라미터 탐색 공간을 정의합니다.
    """
    params = {
        # iterations: 부스팅 라운드(트리)의 수
        'iterations': trial.suggest_int('iterations', 100, 2000, step=100),
        
        # learning_rate: 학습률
        'learning_rate': trial.suggest_loguniform('learning_rate', 1e-3, 0.1),
        
        # depth: 트리의 최대 깊이
        'depth': trial.suggest_int('depth', 4, 10),
        
        # l2_leaf_reg: L2 규제 파라미터
        'l2_leaf_reg': trial.suggest_loguniform('l2_leaf_reg', 1e-8, 10.0),
        
        # subsample: 각 트리를 위한 학습 데이터 샘플링 비율
        'subsample': trial.suggest_uniform('subsample', 0.6, 1.0),
        
        # bagging_temperature: 배깅(bagging)의 강도를 제어
        'bagging_temperature': trial.suggest_loguniform('bagging_temperature', 1e-8, 10.0)
    }
    return params
```

```python
import optuna
import lightgbm as lgb
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from catboost import CatBoostRegressor

def optimize_Regressor_model(X, y, model_name, param_space_fn, n_trials=20, cat_features=None, test_size=0.2, random_state=42):
    # 회귀 문제이므로 stratify=y를 제거합니다.
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    def objective(trial):
        params = param_space_fn(trial)

        if model_name == 'lgbm':
            # lightgbm은 범주형 변수를 명시적으로 지정할 수 있습니다.
            model = lgb.LGBMRegressor(**params, random_state=random_state)
            model.fit(X_train, y_train, categorical_feature=cat_features)
            preds = model.predict(X_test)
        
        elif model_name == 'xgboost':
            # xgboost는 reg:squarederror를 기본 objective로 사용합니다.
            model = xgb.XGBRegressor(**params, random_state=random_state)
            model.fit(X_train, y_train)
            preds = model.predict(X_test)

        elif model_name == 'random_forest':
            model = RandomForestRegressor(**params, random_state=random_state)
            model.fit(X_train, y_train)
            preds = model.predict(X_test)
        elif model_name == 'catboost': # CatBoost 모델 추가
            model = CatBoostRegressor(**params, random_state=random_state, verbose=0)
            # 범주형 변수를 cat_features에 명시적으로 전달
            model.fit(X_train, y_train, cat_features=cat_features)
            preds = model.predict(X_test)
        else:
            raise ValueError(f"Unsupported model: {model_name}")

        # 회귀 문제이므로 R2_score로 평가합니다.
        # Optuna는 maximize를 기본 방향으로 사용하므로, R2_score는 적합합니다.
        return r2_score(y_test, preds)

    study = optuna.create_study(direction="maximize")
    study.optimize(objective, n_trials=n_trials)

    # 최적 모델 학습
    best_params = study.best_params
    if model_name == 'lgbm':
        best_model = lgb.LGBMRegressor(**best_params, random_state=random_state)
        best_model.fit(X_train, y_train, categorical_feature=cat_features)
    elif model_name == 'xgboost':
        best_model = xgb.XGBRegressor(**best_params, random_state=random_state)
        best_model.fit(X_train, y_train)
    elif model_name == 'random_forest':
        best_model = RandomForestRegressor(**best_params, random_state=random_state)
        best_model.fit(X_train, y_train)
    elif model_name == 'catboost':
        best_model = CatBoostRegressor(**best_params, random_state=random_state, verbose=0)
        best_model.fit(X_train, y_train, cat_features=cat_features)

    preds = best_model.predict(X_test)
    best_score = r2_score(y_test, preds)
    
    return best_model, best_score, best_params
```

```python
best_model, best_r2, best_params = optimize_Regressor_model(
    X, y,
    model_name='random_forest',
    param_space_fn=random_forest_space,
    n_trials=1000, 
)

print("최고 R2:", best_r2)
print("최적 하이퍼파라미터:")
print(best_params)
```
