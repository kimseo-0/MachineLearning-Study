# 군집 모델

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
X = iris.data
```

## K Mean

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3, random_state=42)
model.fit(X)
print("클러스터 레이블:", model.labels_)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) # 줄이고 싶은 차원의 수 작성
X_pca = pca.fit_transform(X)

X_pca
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=model.labels_, cmap='rainbow')
plt.title("KMeans 군집 결과 (Iris 데이터)")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.colorbar(label="Cluster")
plt.show()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=0.5, min_samples=5)
labels = model.fit_predict(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='rainbow')
plt.title("DBSCAN 군집 결과")
plt.show()
```
