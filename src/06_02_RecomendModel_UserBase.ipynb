```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
import pandas as pd
pd.set_option('max_colwidth', 100)
```

```python
ratings = pd.read_csv("data/movie2/ratings.csv")
movies = pd.read_csv("data/movie2/movies.csv")
```

```python
ratings.head()
```

```python
movies.head()
```

```python
movie_ratings = pd.merge(
    left = ratings,
    right = movies,
    on = 'movieId'
)
movie_ratings.head()
```

```python
movie_ratings[movie_ratings['userId']==1]
```

```python
title_user = movie_ratings.pivot_table(index='userId', columns='title', values='rating')
title_user
```

```python
title_user = title_user.fillna(0)
title_user
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_based_result = cosine_similarity(title_user, title_user)
user_based_result
```

```python
user_based_df = pd.DataFrame(user_based_result)
user_based_df.head()
```

```python
user_based_df.iloc[0].sort_values(ascending=False)
```
